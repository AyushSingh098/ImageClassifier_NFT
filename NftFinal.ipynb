```python
import cv2
from PIL import Image
import numpy as np
import pandas as pd
import csv
import glob
import os

os.chdir(r"C:\Users\Anand\Desktop\NFT_dataset")

def rename_reshape():
    i=1
    for f in os.listdir():
        img=Image.open(f)
        img=img.resize((300,200),Image.ANTIALIAS)
        img.save(str(i).zfill(3)+'.jpeg',"JPEG")
        i=i+1
        
def create_csv():
    mydata=list()
    for f in os.listdir():
        x,y=os.path.splitext(f)
        if int(x.lstrip('0'))<=5:
            mydata.append([f.strip(),1])
        else:
            mydata.append([f.strip(),-1])
    with open("train_temp.csv","w", newline = '') as myfile:
        wr=csv.writer(myfile)
        wr.writerows(mydata)
    df = pd.read_csv(r"C:\Users\Anand\Desktop\NFT_dataset\train_temp.csv", header=None)
    df.rename(columns={0: "Image", 1: "label"}, inplace=True)
    df.to_csv(r"C:\Users\Anand\Desktop\NFT_dataset\train.csv", index=False) # save to new csv file
    
def create_input_vector():
    inp_X = []
    for img in glob.glob(r"C:\Users\Anand\Desktop\NFT_dataset\*.jpeg"):
        n = cv2.imread(img,0)
        #cv_img.append(n)
        X = np.array(n)
        X = X.flatten()
        X = X[:10]
        print("The image array:")
        print(X)
        inp_X.append(X)
    return inp_X

def create_target_vector():
    colnames = ["Image","label"]
    data = pd.read_csv(r"C:\Users\Anand\Desktop\NFT_dataset\train.csv", names=colnames)
    #images = data.Image.tolist()
    labels = data.label.tolist()
    #images = images[1:]
    labels = labels[1:]
    d = []
    for i in labels:
        d.append(int(i))
    return d
    
class Perceptron(object):
    def __init__(self,input_size,lr,epochs):# Constructor in python
        self.W=np.zeros(input_size+1)# self in python denotes the class atrribute i.e it is data member of the class
        self.lr=lr
        print(self.W)
        self.epochs=epochs
    def activation_fn(self, x):
        if x >= 0:
            return 1
        else:
            return -1
    def predict(self, x):
        #x = np.insert(x, 0, 1)#insert capital B : x has all xi and ,
        z = self.W.T.dot(x)#       Dot product with W i.e weight vectors wi=0 and b=0
        print("The input vector:")
        print(x)
        print("The weight vector:")
        print(self.W)
        print("The dot product of weight vector and input vector:")
        print(z)
        z = np.sum(z)
        print("The final sum:")
        print(z)
        a = self.activation_fn(z)
        return a
    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(len(d)):#d is our target vector d.shape[0] tells us the number of rows
                x = np.insert(X[i], 0, 1)#insert capital B : x has all xi and ,
                print("Input array")
                print(x)
                y = self.predict(x)
                e = d[i] - y
                self.W = self.W + self.lr * e * x
                print("The weight vector after updation:")
                print(self.W)
    def testing(self,X,d):
        count = int()
        count = 0
        for i in range(len(d)):#d is our target vector d.shape[0] tells us the number of rows
            x = np.insert(X[i], 0, 1)#insert capital B : x has all xi and ,
            print("Input array")
            print(x)
            y = self.predict(x)
            print("The target",d[i])
            print("The value obtained is",y)
            if y == d[i]:
                count = count + 1
            print("The weight vector after updation:")
            print(self.W)
        return count
        
# def rename_reshape():
    
# def create_csv():
    
# def create_input():
    
# def create_target():
    
if __name__ == '__main__':
    #img = cv2.imread(r"C:\Users\Anand\Desktop\Nft\004.jpeg", 0)
    #img2 = Image.open(r"C:\Users\Anand\Desktop\Nft\001.jpeg")
    #print(img)
#     Y = np.array(img)
#     Y = Y.flatten()
#     Y = Y[:10]
#     print(Y)
    #x,y = img2.size
    #print(x)
    #print(y)
    #for i in range(0,x):
     #   for j in range(0,y):
      #      print(img[i][j])

    #for d in csv.DictReader(open(r"C:\Users\Anand\Desktop\Nft\trail.csv"), delimiter=','):
     #   print(d)
    #"""df = pd.read_csv(r"C:\Users\Anand\Desktop\Nft\Final.csv")
    #images= df["Image"] #you can also use df['column_name']
    #labels=df["label"]
    #print(images)
    #print(labels)"""
#     colnames = ["Image","label"]
#     data = pd.read_csv(r"C:\Users\Anand\Desktop\Nft\Final.csv", names=colnames)
#     images = data.Image.tolist()
#     labels = data.label.tolist()
#     images = images[1:]
#     labels = labels[1:]
   # print(images)
   # print(labels)
#inp_X = []
   #cv_img = []
#     for img in glob.glob(r"C:\Users\Anand\Desktop\Nft\*.jpeg"):
#         n = cv2.imread(img,0)
#         #cv_img.append(n)
#         X = np.array(n)
#         X = X.flatten()
#         X = X[:10]
#         print("The image array:")
#         print(X)
#         inp_X.append(X)
#     for f in os.listdir():
#         img = cv2.imread(f,0);
#         X = np.array(img)
#         X = X.flatten()
#         X = X[:10]
#         print("The image array:")
#         print(X)
#         inp_X.append(X)
#     print(inp_X)
#     print(inp_X[0])
#     inp=[]
#    for i in inp_X:
#         inp.append(i[0].astype('int32'))
#     print(inp)
#     #print(inp_X)
#     #X = np.array(img)
#     #X = X.flatten()
#     #X = X[:10]
#     print(X)
#     print(X.ndim)
#     print(X.size)
#     d = np.array([0])
#     d = []
#     for i in labels:
#         d.append(int(i))
    rename_reshape()
    create_csv()
    inp = create_input_vector()
    size = int()
    size = len(inp)
    t = create_target_vector()
    size = size//4
    size = size*3
    print(size)
    inp_train = inp[:size]
    t_train = t[:size]
    inp_test = inp[size:]
    t_test = t[size:]
    print("Training",inp_train,t_train)
    print("Testing",inp_test,t_test)
    perceptron = Perceptron(inp_train[0].size,0.1,100)
    perceptron.fit(inp_train, t_train)
    print(perceptron.W)
    count = float()
    count=perceptron.testing(inp_test, t_test)
    print("The total count",count)
    accuracy = float()
    accuracy = (count/size)*100
    print("The accuracy is:",accuracy)
#def testing()
```
